Graficar la siguiente matriz de confusión como un mapa de calor:
|            | Predicted 0 | Predicted 1 |
|------------|-------------|-------------|
| **Actually 0** | 87          | 16          |
| **Actually 1** | 17          | 59          |

El mapa de calor debe lucir algo similar al siguiente gráfico:

![kiku](img/cm2.png)

Se debe utilizar las librerías matplotlib y seaborn.

```python
# Este gráfico ha sido generado reutilizando y modificando el código propuesto
# por el suario 'bninopaul' en stackoverflow disponible en el siguiente enlace:
# https://stackoverflow.com/a/35572247

import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

#Definición de la lista que simula la matriz de confusión a graficar.
confusionMatrix = [[87,16],[17,59]]

# accuracy = (TP + TN) / (TP + FP + TN + FN)
accurracyScore = (confusionMatrix[0][0] + confusionMatrix[1][1]) / (confusionMatrix[0][0] + confusionMatrix[0][1] + confusionMatrix[1][0] + confusionMatrix[1][1])
print(confusionMatrix[0][0])
 
plt.style.use('seaborn')
plt.figure(figsize = (10,5))

# Crear el Dataframe a partir de la lista multidimensional
# almacenanda en la variable 'confusionMatrix'. Se agregan
# los índices y las columnas (False, True)
df = pd.DataFrame(confusionMatrix, index = ['False', 'True'],
                  columns = ['False', 'True'])

# Graficar el mapa de calor con el dataframe creado en la línea anterior,
# En la escala de azules y tamaño de fuente 14 (texto dentro de las celdas)
sn.heatmap(df, annot=True,
                  cmap='Blues', 
                  annot_kws={"fontsize":14}
            )
# Establece el título, concatenando la variable 'accurracyScore' 
plt.title('Accuracy Score: ' + str(accurracyScore), fontsize=16)
plt.xlabel('Predicción', fontsize=14)
plt.ylabel('Etiqueta Correcta', fontsize=14)
```
