```python
# Importar librerías
import numpy as np
import pandas as pd
from sklearn.datasets import load_boston
```

```python
data = load_boston()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df.head()
```

```python
# Establecer semilla para la secuencia de números asleatorios
# Esto garantiza que el experimento se pueda realizar bajo las 
# mismas condiciones, por lo tanto, sea completamente replicable 

# np.random.seed(100)

for _ in range(1, 15):
    samples = np.random.choice(a=15, size=15, replace=True)
    print(samples, ',Num Samples: ' + str(len(samples)) + ',numberUnique: ' + str(len(set(samples))) )
```

```python
bootstrappedDF = df.iloc[samples, :]
bootstrappedDF
```

```python
for _ in range(1, 15):
    samples = np.random.choice(a=15, size=15, replace=False)
    print(samples, ',Num Samples: ' + str(len(samples)) + ',numberUnique: ' + str(len(set(samples))) )
```

```python
withoutDF = df.iloc[samples, :]
withoutDF.iloc[samples, :]
```
